```python
#importing libraries
import jieba
import numpy
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
```

```python
#loading data
data = pd.read_csv("FullTrain.csv")
```

```python
#reading the data
chineseColumn1 = data.title1_zh.to_string()
chineseColumn2 = data.title2_zh.to_string()
```

```python
chineseSegGen = jieba.cut(chineseColumn1, cut_all=False)
chineseSegGen1 = jieba.cut(chineseColumn2,cut_all=False)
```

```python
#Converting generator type to list 
chineseSegList = list(chineseSegGen)
```

```python
chineseSegList1 = list(chineseSegGen1)
```

```python
#removing blank spaces
while ' ' in chineseSegList:
    chineseSegList.remove(' ')
```

```python
#removing blank spaces
while ' ' in chineseSegList1:
    chineseSegList1.remove(' ')
```

```python
yvectorizer = CountVectorizer()
zvectorizer = CountVectorizer()
```

```python
y= yvectorizer.fit_transform(chineseSegList)
z = zvectorizer.fit_transform(chineseSegList1)
```

```python
yfeatures = yvectorizer.get_feature_names()
zfeatures= zvectorizer.get_feature_names()
```

```python
print(yfeatures)
```

```python
print(zfeatures)
```

```python
print('Vocab count y :',len(yfeatures))
print('Number of words y:',len(chineseSegList))
```

```python
print('Vocab count z :',len(zfeatures))
print('Number of words z:',len(chineseSegList1))
```

```python
#converting sparse matrix to an Array
yarrayMatrix = y.toarray()
zarrayMatrix = z.toarray()
```

```python
#converting array to a list
yarrayList=yarrayMatrix.tolist()
zarrayList=zarrayMatrix.tolist()
```

```python
#printing all the elements of counter vector
print(yarrayList)
```

```python
#printing all the elements of counter vector
print(zarrayList)
```
